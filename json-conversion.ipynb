```python
import re 

def to_dict(_object):
    new_obj = {}
    wait = []
    #creates dict object 
    for line in _object:
        line = line.strip()
        #print(line)
        if "Unknown" in line or "jpg" in line or len(line) == 0: #change #$ regular expressions #use tester 
            continue 
        else:
            splt_s = line.split(':')
            if splt_s[0] != " " and splt_s[0] != "": #if empty line, skip 
                for count,item in enumerate(splt_s):
                    splt_s[count] = item.strip()
            new_obj[splt_s[0]] = splt_s[1]

    return new_obj

curr_obj = []
dict_objs = []
count = 0 
for line in lines:
    x = re.search('^Maker', line) #change 
    count = count + 1 
    if line == '\n':
        continue
    elif x!= None and len(curr_obj) != 0:
        dict_objs.append(to_dict(curr_obj))
        curr_obj = []
        curr_obj.append(line)
    elif count == len(test_lines):
        curr_obj.append(line)
        dict_objs.append(to_dict(curr_obj))
    else:
        curr_obj.append(line)
        
        


for i in dict_objs:
    print(i)
    print('\n')
    
json = json.dumps(dict_objs)
    
print(json)
        
    
```
